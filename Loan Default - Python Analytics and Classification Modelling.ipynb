```python
#x = df[["loan_id","monthly_income","origination_score_band","TOB_months","closing_principal_balance","original_loan_amount","product","original_loan_term","remaining_loan_term","delq_history"]].values
#x = x.reshape(-1,7)
x = df[["loan_id", "monthly_income", "origination_score_band", "TOB_months", "closing_principal_balance", "original_loan_amount", "original_loan_term", "remaining_loan_term"]].values
x
```

```python
y = df["target"].values   
y  
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression 
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=20)
```

```python
#Logistic Regression Classification, KNN Classification, Decision Tree Classification,Random Forest Classification,
#Support Vector Machine(SVM), Naive Bayes Classification and XG-Boost.
```

```python
#Logistic Regression Classification Model

model = LogisticRegression()
model_fit = model.fit(X_train,y_train)  
predict = model.predict(X_test)  
predict
```

```python
model.score(X_train,y_train) 
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import mean_squared_error 
```

```python
accuracy_score(y_test,predict) * 100  
```

```python
#TP, FP
#FN, TN
cm = confusion_matrix(y_test,predict)
cm
```

```python
f, ax = plt.subplots(figsize = (8, 8))
sns.heatmap(cm, annot = True, linewidths=1, linecolor="black", fmt =".0f", ax=ax)
plt.xlabel("predict")
plt.ylabel("y_test")
plt.show()
```

```python
#Decided to go with the Random Forest Classification Model having tried some other models including 
# Linear, Decision Tree, KNN, SVC and XGBoost Classification Models.

#Using Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
# set n_estimators to 100 whcih means the model will use 100 subsets
#rf = RandomForestClassifier(n_estimators = 100, random_state = 20)

rf = RandomForestClassifier()
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=20)
rf.fit(X_train, y_train)
print("random forest model score: ", rf.score(X_test, y_test))
rf_score= rf.score(X_test, y_test)
```

```python
#Accuracy Score for random forest

y_predict = rf.predict(X_test)
accuracy = accuracy_score(y_test, y_predict)
accuracy
```

```python
#Confusion matrix for random forest

y_predict = rf.predict(X_test)
y_actual = y_test
cm = confusion_matrix(y_actual, y_predict)
cm
```

```python
#Heatmap for random forest

f, ax = plt.subplots(figsize=(8,8))
sns.heatmap(cm, annot = True, linewidths=1, linecolor = "green", fmt = ".0f", ax=ax)
plt.xlabel("y_predict")
plt.ylabel("y_actual")
plt.show()
```

```python
#Accuracy Score for KNeighborsClassifier

from sklearn.neighbors import KNeighborsClassifier
k=11
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train, y_train)
prediction = knn.predict(X_test)
print("{} nn score: {}".format(k, knn.score(X_test, y_test)))

knn_score = knn.score(X_test, y_test)
```

```python
#Accuracy Score for Support Vector Machine(SVM)

from sklearn.svm import SVC
svm = SVC(random_state = 42)
svm.fit(X_train, y_train)

print("Accuracy of SVM: ", svm.score(X_test, y_test))

svm_score = svm.score(X_test, y_test)
```

```python
#Accuracy Score for Decision Tree Classification

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state = 20)
dt.fit(X_train, y_train)

print("score: ", dt.score(X_test, y_test))
dt_score = dt.score(X_test, y_test)
y_prediction = dt.predict(X_test)
y_actual = y_test
cm = confusion_matrix(y_actual, y_prediction)
```
